```python
import optuna
from optuna import Trial
from optuna.samplers import TPESampler
from sklearn.model_selection import cross_val_score

def objective(trial: Trial) -> float:
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 500, 5000),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'subsample': trial.suggest_float('subsample', 0.5, 1),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1),
        'gamma': trial.suggest_float('gamma', 0, 1),
        'random_state': 42,
        'n_jobs': -1
    }
    model = XGBRegressor(**params)
    return cross_val_score(model, X_clean, y_train, n_jobs=-1, cv=5, scoring='neg_mean_squared_error').mean()

sampler = TPESampler(seed=42)
study = optuna.create_study(direction='maximize', sampler=sampler)
study.optimize(objective, n_trials=100) 
```

```python
# on recup le meilleur parametre
best_params = study.best_params
best_params['random_state'] = 42
best_params['n_jobs'] = -1
# entrainement avec best_params
best_params
```

```python
X = pd.read_csv('data/X_train_6ZIKlTY.csv')
y = pd.read_csv('data/y_train_lXj6X5y.csv')
data = PreprocessData(X, y)
```

```python
X1 = data.get_X()
y1 = data.get_y()
X1.head()
```

```python
X1_clean = X1.drop(['ledd','off','time_since_intake_on','time_since_intake_off'],axis=1)
```

```python
X1_clean.head()
```

```python
# recuperer les lignes ou il y a un NaN sur la colonne 'on'
X1_a_imputer = X1_clean[X1_clean['on'].isnull()]

X1_sans_nan = X1_clean.dropna()
X1_sans_nan.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1_sans_nan.drop(['on'],axis=1), X1_sans_nan['on'], test_size=0.2, random_state=42)
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
# Entraînement avec early stopping
model = xgb.XGBRegressor(n_estimators=3000, learning_rate=0.03,     eval_metric='rmse',  # Spécifier la métrique ici
random_state=42,          early_stopping_rounds=10,
)
model.fit(X_train, y_train, 
          eval_set=[(X_train, y_train), (X_test, y_test)],
          verbose=True)

# Afficher la meilleure itération
print(f"Meilleure itération: {model.best_iteration}")
print(f"Meilleur score: {model.best_score}")

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import explained_variance_score, median_absolute_error

# Faire les prédictions
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Calculer diverses métriques d'évaluation
metrics = {
    'MSE (train)': mean_squared_error(y_train, y_pred_train),
    'MSE (test)': mean_squared_error(y_test, y_pred_test),
    'RMSE (train)': np.sqrt(mean_squared_error(y_train, y_pred_train)),
    'RMSE (test)': np.sqrt(mean_squared_error(y_test, y_pred_test)),
    'MAE (train)': mean_absolute_error(y_train, y_pred_train),
    'MAE (test)': mean_absolute_error(y_test, y_pred_test),
    'R² (train)': r2_score(y_train, y_pred_train),
    'R² (test)': r2_score(y_test, y_pred_test),
    'Explained Variance (train)': explained_variance_score(y_train, y_pred_train),
    'Explained Variance (test)': explained_variance_score(y_test, y_pred_test),
    'Median Absolute Error (train)': median_absolute_error(y_train, y_pred_train),
    'Median Absolute Error (test)': median_absolute_error(y_test, y_pred_test)
}

# Afficher les métriques
for metric_name, metric_value in metrics.items():
    print(f"{metric_name}: {metric_value:.4f}")

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_test, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Valeurs réelles')
plt.ylabel('Valeurs prédites')
plt.title('Valeurs prédites vs Valeurs réelles (Test)')
plt.grid(True)
plt.show()

residuals_test = y_test - y_pred_test
plt.figure(figsize=(10, 6))
plt.hist(residuals_test, bins=30, alpha=0.7, color='blue')
plt.axvline(0, color='red', linestyle='--')
plt.xlabel('Résidus')
plt.ylabel('Fréquence')
plt.title('Distribution des résidus (Test)')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(y_pred_test, residuals_test, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Valeurs prédites')
plt.ylabel('Résidus')
plt.title('Résidus vs Valeurs prédites (Test)')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
xgb.plot_importance(model, importance_type='weight', max_num_features=20)
plt.title('Importance des features')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(y_test.reset_index(drop=True), 'b-', label='Valeurs réelles')
plt.plot(y_pred_test, 'r-', label='Valeurs prédites')
plt.xlabel('Indice')
plt.ylabel('Valeur')
plt.title('Valeurs prédites et réelles par indice (Test)')
plt.legend()
plt.grid(True)
plt.show()

if hasattr(model, 'evals_result'):
    results = model.evals_result()
    if results:
        plt.figure(figsize=(10, 6))
        epochs = len(results['validation_0']['rmse'])
        x_axis = range(0, epochs)
        plt.plot(x_axis, results['validation_0']['rmse'], label='Train')
        plt.plot(x_axis, results['validation_1']['rmse'], label='Test')
        plt.legend()
        plt.xlabel('Itérations')
        plt.ylabel('RMSE')
        plt.title('XGBoost RMSE')
        plt.grid(True)
        plt.show()
```
